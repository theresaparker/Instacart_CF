## Loyalty Flag

#### Question 4: Create a loyalty flag for existing customers using the transform and loc() function. 

```python
# creating new column for max_orders per customer
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# checking output
df.head()
```

using the loc() function to group customers

```python
# using the loc() function to group customers who have over 40 orders as "Loyal Customer"
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
```

```python
# using the loc() function to group customers who have less than 40, but more than 10 as 'Regular Customer'
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
# using the loc() function to group customers who have 10 or less orders as 'New Customer'
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
# cheking output/frequencies of flags
df['loyalty_flag'].value_counts(dropna = False)
```

```python
# viewing output
df.head()
```

### Question 5: Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). Note: What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

## Spending Flag

#### Question 6: Create a spending flag for each user based on the average price across all their orders using the following criteria:
* If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
* If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# creating a new column for average prices for each user
df['avg_price']=df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df.head()
```

```python
# creating a flag for customers whose average order price is is great than 10 using loc()
df.loc[df['avg_price'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
# creating a flag for customers whose average order price is less than 10 using loc()
df.loc[df['avg_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
# checking output
df['spending_flag'].value_counts(dropna=False)
```

```python
df.head()
```

## Frequency Flag

#### Question 7: Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
* If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
* If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
* If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# creating new column for median order duration
df['median_order_duration'] = df.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# creating a flag for customers whose median order duration is greater than 20 as 'Non-Frequent Customer'
df.loc[df['median_order_duration'] > 20, 'frequency_flag'] = 'Non-Frequent Customer'
```

```python
# creating a flag for customers whose median order duration is greater than 10 but less than/equal to 20 as 'Regular Customer'
df.loc[(df['median_order_duration'] > 10) & (df['median_order_duration'] <= 20), 'frequency_flag'] = 'Regular Customer'
```

```python
# creating a flag for customers whose median order duration is less than or equal to 10 as 'Frequent Customer'
df.loc[df['median_order_duration'] <= 10, 'frequency_flag'] = "Frequent Customer"
```

```python
df['frequency_flag'].value_counts(dropna=False)
```

#### Customers who have a median order duration of 10 or less are the majority of customers.
