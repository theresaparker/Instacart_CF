### Alternative method to subsetting using the loc function
iloc looks for integer position; 
loc looks for a column name

```python
# Using (.ISIN) to locate more than one value. Department_id's of 19, 20, 21. 

df_snacks_2 = df_prods.loc[df_prods['department_id'].isin([19,20,21])]
```

```python
df_snacks_2
```

```python
# Using (==) 19 to pull department 19
df_snacks_2 = df_prods.loc[df_prods['department_id']==19]
```
